**Merging Rainfall Data**

```python
rain_df = pd.read_csv('https://raw.githubusercontent.com/ManjuVijayakumar/ICTAK_DSA_FinalProject/refs/heads/main/Sub_Division_IMD_1997_2017.csv')
rain_df.sample(3)
```

```python
rain_df = rain_df.rename(columns={"SUBDIVISION": "Subdivision", "YEAR": "Year"})
```

```python
mapping = {
    "ANDAMAN & NICOBAR ISLANDS": ["Andaman and Nicobar Islands"],
    "ARUNACHAL PRADESH": ["Arunachal Pradesh"],
    "ASSAM & MEGHALAYA": ["Assam", "Meghalaya"],
    "NAGA MANI MIZO TRIPURA": ["Nagaland", "Manipur", "Mizoram", "Tripura"],
    "SUB HIMALAYAN WEST BENGAL & SIKKIM": ["West Bengal", "Sikkim"],
    "GANGETIC WEST BENGAL": ["West Bengal"],
    "ORISSA": ["Odisha"],
    "JHARKHAND": ["Jharkhand"],
    "BIHAR": ["Bihar"],
    "EAST UTTAR PRADESH": ["Uttar Pradesh"],
    "WEST UTTAR PRADESH": ["Uttar Pradesh"],
    "UTTARAKHAND": ["Uttarakhand"],
    "HARYANA DELHI & CHANDIGARH": ["Haryana", "Delhi", "Chandigarh"],
    "PUNJAB": ["Punjab"],
    "HIMACHAL PRADESH": ["Himachal Pradesh"],
    "JAMMU & KASHMIR": ["Jammu and Kashmir"],
    "WEST RAJASTHAN": ["Rajasthan"],
    "EAST RAJASTHAN": ["Rajasthan"],
    "WEST MADHYA PRADESH": ["Madhya Pradesh"],
    "EAST MADHYA PRADESH": ["Madhya Pradesh"],
    "GUJARAT REGION": ["Gujarat"],
    "SAURASHTRA & KUTCH": ["Gujarat"],
    "KONKAN & GOA": ["Goa", "Maharashtra"],
    "MADHYA MAHARASHTRA": ["Maharashtra"],
    "MATATHWADA": ["Maharashtra"],
    "VIDARBHA": ["Maharashtra"],
    "CHHATTISGARH": ["Chhattisgarh"],
    "COASTAL ANDHRA PRADESH": ["Andhra Pradesh"],
    "TELANGANA": ["Telangana"],
    "RAYALSEEMA": ["Andhra Pradesh"],
    "TAMIL NADU": ["Tamil Nadu"],
    "COASTAL KARNATAKA": ["Karnataka"],
    "NORTH INTERIOR KARNATAKA": ["Karnataka"],
    "SOUTH INTERIOR KARNATAKA": ["Karnataka"],
    "KERALA": ["Kerala"],
    "LAKSHADWEEP": ["Lakshadweep"]
}
```

```python
rain_df.head(2)
```

```python
expanded_rows = []
for _, row in rain_df.iterrows():
    states = mapping.get(row["Subdivision"], [])
    for st in states:
        new_row = row.copy()
        new_row["State_rainfall"] = st
        expanded_rows.append(new_row)

rain_expanded = pd.DataFrame(expanded_rows)
```

```python
rain_expanded.head(3)
```

```python
data = pd.merge(
    data,
    rain_expanded,
    left_on= ['State', 'Crop_Year'],
    right_on=["State_rainfall", "Year"],
    how="left"
    ).drop(columns= ['Subdivision', 'Year', 'State_rainfall'])

# Annual_Rainfall
# ANNUAL_rain
```

```python
data[data['Annual_Rainfall'] != data['ANNUAL_rain']]
```

```python
# Update annual rainfall from rain_expanded incase different

data['Annual_Rainfall'] = data['ANNUAL_rain'].combine_first(data['Annual_Rainfall'])
data.drop(columns=['ANNUAL_rain'], inplace= True)
```
